```python
paragraph = "THE fat CAT jump over the lazy dog dog and the mouse."
print(paragraph)
```

```python
print(paragraph.lower())  # Convert the entire string to lowercase
print(paragraph.upper())  # Convert the entire string to uppercase
```

```python
print(paragraph.title())  # Convert the first character of each word to uppercase
```

```python
print(paragraph.capitalize())  # Convert the first character of the string to uppercase
```

```python
words = paragraph.lower().split()  # Split the string into a list of words
print(words)
for i in words:
    if i == "mouse":
        print("Found the word 'mouse'!")
    else:
        print("'mouse' is not the word.")
```

```python
from collections import Counter


words = paragraph.lower().split()

w = Counter(words)  # Count the occurrences of each word in the list

print(w)
```

```python
w.items()  # Get the items (word, count) as a list of tuples
```

```python
for word, count in w.items():
    if count > 1:
        print(f"The word '{word}' appears {count} times.")
```

```python
person = {"name": "Alice", "age": 30, "is_student": False, "currency": "USD", "height": 1.75}

print(person)
```

```python
print(person.items())  # Get the items (key, value) as a list of tuples
```

```python
person.keys()  # Get the keys of the dictionary
```

```python
person.values()  # Get the values of the dictionary
```

```python
for key, value in person.items():
    print(f"{key}: {value}")
```

```python
for key, value in person.items():
    if value == 'Alice':
        print(value)
        print(f"Found the key 'name' with value 'Alice'!")
```

```python
# dictionary functions 

person.get("names", "Key not found")  # Get the value associated with the key "name"
```

```python
person['name']  # Get the value associated with the key "name" using square brackets
```

## Zip

```python
names = ["alice", "john", "peter"]
ages = [30, 60, 99]

for age, name in zip(ages, names):
    print(f"{name} is {age} years old.")
```

```python
for _, name in zip(ages, names):
    print(f"{name}")  # Using underscore for unused variable
```

```python
for age, _ in zip(ages, names):
    print(f"{age}")  # Using underscore for unused variable
```

## Nested for loops

```python
names = ["alice", "john", "peter"]
ages = [30, 60, 99]
heights = [1.75, 1.80, 1.65]

for age in ages:
    for name in names:
        for height in heights:
            print(f"{name} is {age} years old and {height} meters tall.")
```

```python
stock_data = [56.2, 'nan', 51.7, 55.3, 42.6, 'nan', 67.8]

stock_data[1] = 0.0  # Replace 'nan' with 0.0
stock_data[5] = 0.0  # Replace 'nan' with 0.0

print(stock_data)
```

```python
for index in range(0, len(stock_data)):
    if stock_data[index] == 'nan':
        stock_data[index] = 0.0  # Replace 'nan' with 0.0

print(stock_data)
```
