## Nested if-else conditions

```python
yahoo = int(input("Enter the Yahoo stock value: "))
if yahoo > 10:
    print("Yahoo stock is doing well today!")
    if yahoo % 2 == 0:
        print("Yahoo stock is an even number.")
    else:
        print("Yahoo stock is an odd number.")
else:
    print("Yahoo stock is not doing well today.")
```

### Checking if an element is in a list

```python
stock_prices = [150, 200, 250, 300, 350]

stock_price_to_check = 300

if stock_price_to_check in stock_prices:
    print(f"The stock price {stock_price_to_check} is in the list.")
else:
    print(f"The stock price {stock_price_to_check} is not in the list.")
```

```python
stock_value = "Yes" if stock_price_to_check in stock_prices else "No"

print(f"Is the stock price {stock_price_to_check} in the list? {stock_value}")
```
