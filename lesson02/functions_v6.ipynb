```python
def visualize_sales(months: list, sales: list):
    plt.plot(month, sales)
    plt.xlabel("months")
    plt.ylabel("sales")
    plt.title("monthly sales trend")
    plt.savefig("sales_trend.png")
    # plt.legend(loc="upper left")
```

```python
month = ["Jan", "Feb", "Mar", "Apr", "May", "June"]
sales = [200, 300, 250, 400, 1500, 550]

visualize_sales(month, sales)
```

```python
visualize_sales(month, sales)
```
